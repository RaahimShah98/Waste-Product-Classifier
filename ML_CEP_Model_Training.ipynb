```python
import tensorflow as tf
```

```python

def convert_to_tensor(file_name , img_size = 224):
  ## Read the image file
  image = tf.io.read_file(file_name)

  ## Decode the already read file into tensors with channels (R,G,B)
  image = tf.image.decode_jpeg(image , channels = 3)

  ## Normalizing images from 0-255 to 0-1
  image  = tf.image.convert_image_dtype(image , tf.float32)
  ## Resizing the image so all images are of same dimensions
  image = tf.image.resize(image , size = [img_size , img_size])

  return image
```

```python
## Gives a tuple of image with its corresponding label
def get_image_label(image_path  , label):
  image = convert_to_tensor(image_path)

  return image , label
```

```python
## defining batch size
BATCH_SIZE = 32

## Creating BATCHES
def create_data_bathces(x , y=None , batch_size=BATCH_SIZE , valid_data=False , test_data  = False):
  if test_data:
    print("Creating Test Data Batch....")
    data = tf.data.Dataset.from_tensor_slices((tf.constant(x) , tf.constant(y)))
    data_batch = data.map(get_image_label).batch(batch_size)
    return data_batch

  elif valid_data:
    print("Creating Valid Data Batch....")
    data = tf.data.Dataset.from_tensor_slices((tf.constant(x) , tf.constant(y)))
    data_batch = data.map(get_image_label).batch(batch_size)
    return data_batch

  else:
    print("Creating Training Data Batch....")
    data = tf.data.Dataset.from_tensor_slices((tf.constant(x) , tf.constant(y)))
    ## Shuffling the Training DataSet
    data = data.shuffle(buffer_size = len(x))

    data_batch = data.map(get_image_label).batch(batch_size)
    return data_batch
```

## Setting up our Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
np.shape(unique_labels)[0]
```

```python
input_shape = (224, 224, 3)  # Adjust based on your image size
num_classes = np.shape(unique_labels)[0]  # Replace with the number of waste material categories

# Define the custom CNN model
def create_model(convo1 = 32 , convo2 = 64, convo3 = 128 , drop_out1 = 0.3 , drop_out2 = 0.3 ,drop_out3 = 0.3 ,drop_out4 = 0.5 , dense = 512 , convo_filter = (3,3)):
    model = Sequential([
        # First convolutional block
        Conv2D(convo1, (3, 3), activation='relu', input_shape=input_shape),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(drop_out1),

        # Second convolutional block
        Conv2D(convo2, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(drop_out2),

        # Third convolutional block
        Conv2D(convo3, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D(pool_size=(2, 2)),
        Dropout(drop_out3),

        # Flatten and fully connected layers
        Flatten(),
        Dense(dense, activation='relu'),
        Dropout(drop_out4 ),
        Dense(num_classes, activation='softmax')
    ])

    return model
```

```python
model = create_model()

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',  # Use sparse_categorical_crossentropy if labels are not one-hot encoded
    metrics=['accuracy']
)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.callbacks import EarlyStopping

checkpoint_callback = ModelCheckpoint(
    filepath= f"/content/drive/MyDrive/University/CEP ML/Models/CNN-Model-1-3.keras",
    # save_weights_only=True,  # Saves only the model's weights
    monitor='val_loss',      # Monitor the validation loss
    mode='min',              # Mode to select the best model (minimizing val_loss)
    save_best_only=True,     # Save only the best model based on the monitored metric
    verbose=1                # Print a message when saving the model
  )

early_stopping = EarlyStopping(monitor='val_loss', patience=3 , restore_best_weights=True)
```
